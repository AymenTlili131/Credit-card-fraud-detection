```python
import pandas as pd
dfTrain= pd.read_csv('../fraudTrain.csv')
dfTest= pd.read_csv('../fraudTest.csv')
```

```python
dfTrain.shape , dfTest.shape
```

```python
df = pd.concat([dfTrain, dfTest], axis=0)
```

```python
df.shape
```

```python
len(df)/4
```

```python
df1=df.iloc[0:int(len(df)/4),:]
df2=df.iloc[int(len(df)/4)+1:int(2*len(df)/4),:]
df3=df.iloc[int(2*len(df)/4)+1:int(3*len(df)/4),:]
df4=df.iloc[int(3*len(df)/4)+1:,:]
```

```python
df1.shape,df2.shape,df3.shape,df4.shape
```

```python
import os  
os.makedirs('../Data/Batch 1/', exist_ok=True)  
df1.to_csv('../Data/Batch 1/Batch1.csv' ,index=False)  
os.makedirs('../Data/Batch 2/', exist_ok=True)  
df2.to_csv('../Data/Batch 2/Batch2.csv',index=False)  
os.makedirs('../Data/Batch 3/', exist_ok=True)  
df3.to_csv('../Data/Batch 3/Batch3.csv',index=False)  
os.makedirs('../Data/Batch 4/', exist_ok=True)  
df4.to_csv('../Data/Batch 4/Batch4.csv',index=False)  
```

```python
df1OOB=df1.iloc[:int(len(df2)/4),:] 
df1OOB.to_csv('../Data/Batch 1/OutOfBoxTest.csv', index=False) 
```
